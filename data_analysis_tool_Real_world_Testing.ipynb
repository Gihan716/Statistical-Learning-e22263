## Data Import

```python
import pandas as pd
from data_analysis import DataInspector, PlottingMethods

# Initialize the tool
inspector = DataInspector()

print("--- 1. DATA INGESTION ---")
# Using the direct URL from the sample cookbook for seamless testing
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
inspector.df = pd.read_csv(url)

# Run your custom sanitization method to clean junk strings and fix types
inspector._sanitize_data()

# View the initial structure
inspector.data_summary()
```

## Structural Analysis & Cleaning (Impute & Prune)

```python
print("--- 2. DATA CLEANING & IMPUTATION ---")

# Handle missing values (Numeric uses median, Categorical defaults to mode)
inspector.handle_missing_values(strategy='median')

# Remove any accidental duplicates
inspector.remove_duplicates()

# Detect and optionally delete outliers in specific numeric columns using IQR
inspector.handle_outliers(columns=['Age', 'Fare'], find_and_delete=True)

# Verify the changes
inspector.data_summary()
```

## Feature Engineering (Normalize)

```python
print("--- 3. FEATURE ENGINEERING PREPARATION ---")

# Standardize numeric columns (Z-score)
scaled_nums = inspector.extract_normalized_numeric_data(method='standard')
display(scaled_nums.head())

# One-hot encode categorical columns
encoded_cats = inspector.extract_normalized_categorical_data(method='onehot')
display(encoded_cats.head())

# Create the final, machine-learning-ready unified dataset
final_ml_df = inspector.create_normalized_data_df()
display(final_ml_df.head())
```

## Visualize Associations.

```python
print("--- 4. INTERACTIVE VISUALIZATIONS ---")

# 1. 3-Panel Univariate Subplots for continuous numbers
inspector.plot_numerical(column_names=['Age', 'Fare'])

# 2. Categorical Frequencies with percentages
inspector.plot_categorical(column_names=['Survived', 'Sex', 'Pclass'])

# 3. Smart Relationships
# Cat-Num (Generates Box Plot)
inspector.plot_relationship('Survived', 'Age')
# Num-Num (Generates Scatter with OLS)
inspector.plot_relationship('Age', 'Fare')
# Cat-Cat (Generates Grouped Bar)
inspector.plot_relationship('Sex', 'Survived')
```

```python
print("--- 5. DEEP STATISTICAL INSIGHTS ---")

# Generate the unified heatmap (Pearson + Cramér's V + Eta)
# Note: This will calculate associations across all columns, revealing
# strong correlations like 'Sex' vs 'Survived' or 'Pclass' vs 'Fare'.
inspector.plot_all_associations_heatmap()
```

```python
print("--- 6. CUSTOM MODULAR PLOTTING ---")

# Initialize the standalone plotting class
PLT = PlottingMethods()

# Plot 1: Pie Chart showing Passenger Class distribution
pie_result = PLT.plot_pie_chart(
    names='Pclass',
    values='Fare',
    data=inspector.df,
    hole=0.4,
    title='Fare Contribution by Passenger Class'
)
PLT.display_image(pie_result)

# Plot 2: Histogram of Passenger Ages
hist_result = PLT.plot_histogram(
    x='Age',
    data=inspector.df,
    title='Distribution of Passenger Ages'
)
PLT.display_image(hist_result)
```
